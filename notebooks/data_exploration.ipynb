```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
```

```python
dataset = pd.read_csv('../data/dataset.csv')
```

```python
for column in dataset.columns:
    dataset = dataset.dropna(subset = [column])
```

```python
print(dataset.head())
print(dataset.describe())
print(dataset.info())
# Very small amount of sample has missing values, should drop it instead
```

```python
fig, axes = plt.subplots(3, 4, figsize=(20, 15))
axes = axes.flatten()

numerical_columns = ['Age', 'Systolic BP', 'Diastolic', 'BS', 
                     'Body Temp', 'BMI', 'Heart Rate', 
                     ]


for i, column in enumerate(numerical_columns):
    if i < len(axes):
        sns.boxplot(x='Risk Level', y=column, data=dataset, ax=axes[i])
        axes[i].set_title(f'{column} by Risk Level')


for j in range(len(numerical_columns), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
